EVOKE-01에서 트로델비의 전체 생존율은 11.9개월으로 대조군인 Docetaxel 8.9개월 대비 통계적 유의미성을 확보하지 못하였다.\n\n다이이찌산쿄의 Dato-Dxd도 비소세포폐암 적응증에서의 객관적 반응율이 펨브롤리주맙에 대한 결과는 좋지 않았다. 2024년 2월 내평가 비소세포폐암 2차 치료제로만 허가 신청을 완료(국가 2월, 유럽 3월)하였다. 하지만, 최근 전체생존율(OS) 데이터도 12.9개월(vs. 18개월)을 발표하며 통계적 유의성을 확보하지 못하였다.\n\nLCB84는 Trop2 항체에 MMAE(모노메틸 오리스타틴 E; monomethyl auristatin E) 페이로드, 리거켐바이오의 링커로 구성되어있다. 출처된 길리어드의 트로델비는 SN-38의 캠토테신 계열 토포이소머레이즈 저해제를 이용한다. 다이이찌산쿄의 Dato-Dxd도 토포이소머레이즈 저해제이다. 리거켐바이오가 사용하는 페이로드인 MMAE의 독성이 트로델비 및 Dato-Dxd의 페이로드보다 독성이 더 높다. 높은 효율을 기대할 수 있으나 정상세포에 페이로드가 작용했을 때 부작용 문제가 생길 수 있다. 따라서 LCB84 임상 1상에서 안정성 결과가 매우 중요하다.\n\n안정성 결과가 잘 나온다면 이후 트로델비와 Dato-Dxd가 실패한 폐암 적응증에서의 임상 성공에 대한 기대감이 높아질 것으로 보인다. 리거켐바이오의 Trop2 ADC(LCB84)는 2023년 12월 임상 1상 첫환자 투여가 완료예정이다. 또한 기술이전 안테임 2.2조원의 기술이전 되었다. 계약금 1,300억원이 유입되었으며, 1/2상 중 안테임 옵션을 행사하면 마일스톤 수취도 가능하다.\n\nMirae Asset Securities Research\n\n\n\n리거켐바이오\n\n# 그림 1.
